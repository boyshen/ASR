## 数据集

语料数据集来自[thchs-30](http://www.openslr.org/18/)。[thchs-30](http://www.openslr.org/18/) 是由清华大学语音与语言技术中心（CSLT）发布的开放式中文语音数据库。原始记录是2002年由王冬在清华大学计算机科学系智能与系统国家重点实验室朱晓燕教授的监督下进行的，原始名称为“ TCMSD”，代表“清华连续大学”普通话语音数据库”。该数据库对学术用户完全免费。

[thchs-30](http://www.openslr.org/18/) 是在安静的办公室环境下，通过单个碳粒麦克风录取的，总时长超过30个小时。大部分参与录音的人员是会说流利普通话的大学生。采样频率16kHz，采样大小16bits。

加载数据集

```python
def handle(line):
    """
    处理每行文本的函数
    :param line: (str, mandatory) 字符／句子。例如 hello word
    :return: (list) 列表。例如 [hello, world]
    """
    line_data = list()
    for word in line.split(' '):
        # 处理空格字符。发现空格字符则跳过。
        word = re.sub('[ ]+', '', word)
        if word == ' ' or word == '' or len(word) == 0:
            continue

        line_data.append(word)

    return line_data

def load_dataset(source_data_path, data_path):
    """
    加载数据集
    :param data_path: (str, mandatory) 训练数据所在的路径
    :param source_data_path: (str, mandatory) 实际数据路径
    :return: (dict of list) 列表字典。返回音频文件路径，标签文件路径，标签数据路径，拼音字符，汉字
    """
    audio_data, pinyin_data, chinese_data = list(), list(), list()

    data_file = os.listdir(data_path)
    for file in tqdm(data_file):
        # 检查文件名后缀是不是 wav 音频类型文件。
        if file[-3:] == 'wav':
            # 标签文件。根据音频文件名得到标签文件名
            label_file = file + '.trn'

            # 判断标签文件是否存在。不存在则报错
            assert label_file in data_file, FileNotFoundException(label_file)

            # 音频文件路径和标签文件路径
            audio_file_path = os.path.join(data_path, file)
            label_file_path = os.path.join(data_path, label_file)

            # 读取标签文件内容。找到对应标签数据路径
            label_data_file_path = DataUtils.read_text_data(label_file_path, show_progress_bar=False)
            assert len(label_data_file_path) == 1, \
                'Get redundant data path: {}, label_file_path:{}'.format(label_data_file_path, label_file_path)

            # 重新拼接路径
            label_data_file_path = os.path.join(source_data_path, label_data_file_path[0].split('/')[-1])
            assert os.path.isfile(label_data_file_path), FileNotFoundException(label_data_file_path)

            # 读取标签数据。包括字符、拼音
            text_data = DataUtils.read_text_data(label_data_file_path, handle_func=handle, show_progress_bar=False)
            chinese = text_data[0]
            pinyin = text_data[1]

            # 检查pinyin字符是否正确。主要判断每个拼音的最后一个字符是不是数字。不是数字则报错
            for py in pinyin:
                assert py[-1].isdigit(), "The last character:{} of Pinyin is not a number! " \
                                         "pinyin_str:{}, " \
                                         "label_data_file_path:{}, " \
                                         "label_file_path:{}".format(py, pinyin, label_data_file_path,
                                                                     label_file_path)

            # 将由多个中文字符组成的词转换成单个字
            new_chinese = list()
            for ch in chinese:
                new_chinese += list(ch)
            chinese = new_chinese

            # 检查是否是一个拼音对应中文字符， 如果不是则报错
            assert len(chinese) == len(pinyin), "The number of pinyin:{} and chinese:{} is different, " \
                                                 "chinese:{}, pinyin:{}, file:{}".format(len(pinyin), len(chinese),
                                                                                         chinese, pinyin,
                                                                                         label_data_file_path)

            audio_data.append(audio_file_path)
            pinyin_data.append(pinyin)
            chinese_data.append(chinese)

    return audio_data, pinyin_data, chinese_data
```

```python
train_audio, train_pinyin, train_chinese = load_dataset(source_data_path, train_data_path)
test_audio, test_pinyin, test_chinese = load_dataset(source_data_path, test_data_path)
```

### 特征提取

音频特征提取，提供两种的常用的提取方式。声谱图和梅尔频率倒谱系数是提取语音特征的两种方式。主要的区别在于MFCC特征维度比声谱图特征少很多，有助于在训练声学模型时候避免数据过拟合。

（1）声谱图(spectrogram)。 

    声谱图是音频信号在时间上的频域表示法，创建方法是将音频信号拆分为频率分量并绘制出相对时间的变化图。声谱图上任何给定点的颜色强度表示信号的振幅。
    
    参考：Mitch Marcus，“CIS 391 人工智能”，费城 (2015)，Seas.upenn.edu。

（2）梅尔倒谱系数分析(MFCC)。

    梅尔刻度。梅尔刻度于1937年提出，描述了人类可真正识别的音高，对于人类来说，听不到的声音就没有必要留在数据里。 为了提取特征，可以给频谱图的频率分组，分组依据与人耳接收能力有关，这样就能过滤我们所听不见的声音，从而大幅减少需要处理的频率数。简单来说梅尔刻度就是提取人类能识别的声音，抛弃多余信息特征。
    
    倒谱系数。每个人的发音都有不同，可以把人声想成是声源和过滤器的组合体。每个人的声源都不同，过滤器则是说话时用到的单词发音。倒谱分析就靠这个模型来区分两者。可以用算法从信号中提取出来。记住关键点是舍弃带有个体声带特征的语音成分，保留声道形成的那部分声音形状。可以理解为倒谱系数就是忽略每个人的不同发音。
    
    将梅尔频率分析与倒谱分析组合，就可以得到与语音相关的12 或 13 中 MFCC 特征。同时还可以选择在特征集附加多个MFCC特征，虽然附加特征会使特征数增加一倍到两倍，但事实证明这样 ASR 效果更好。 
    
    使用 MFCC 好处就是不仅能大幅度减少数据维数，还能过滤掉系统噪声。

### 包装数据集

使用 tensorflow 提供的 tf.data.Dataset 将数据集进行包装成 tensor 类型，可以加快训练。同时设置 batch_size、shuffle。

```python
import numpy as np
import tensorflow as tf

def packaging_dataset(audio_data, pinyin_data, audio_feature, pinyin_dict, batch_size=2, shuffle=True):
    """
    包装数据集
    :param audio_data: (list, mandatory) 音频数据
    :param pinyin_data: (list, mandatory) 拼音标签数据
    :param audio_feature: (AudioFeature, mandatory) 音频数据特征提取
    :param pinyin_dict: (PinYin, mandatory) 拼音数据提取
    :param batch_size: (int, optional, default=2) 批量样本
    :param shuffle: (bool, optional, default=True) 是否洗牌
    :return: (tf.data.Dataset) 数据集
    """
    x_data, input_data, y_data, label_data = list(), list(), list(), list()
    for a_data, p_data in tqdm(zip(audio_data, pinyin_data)):
        x, input_length = audio_feature.feature(a_data, alignment=True)
        y, label_length = pinyin_dict.encoding(p_data, alignment=True)

        input_length = [input_length]
        label_length = [label_length]

        x_data.append(x.astype(np.float32).tolist())
        input_data.append(input_length)
        y_data.append(y.astype(np.int32).tolist())
        label_data.append(label_length)

    x = tf.convert_to_tensor(x_data, dtype=tf.float32)
    input_length = tf.convert_to_tensor(input_data, dtype=tf.int32)
    y = tf.convert_to_tensor(y_data, dtype=tf.int32)
    label_length = tf.convert_to_tensor(label_data, dtype=tf.int32)

    dataset = tf.data.Dataset.from_tensor_slices((x, input_length, y, label_length))
    if shuffle:
        dataset = dataset.shuffle(batch_size * 2).batch(batch_size)
    else:
        dataset = dataset.batch(batch_size)

    return dataset
```

```python
batch_size = 2

train_dataset = packaging_dataset(train_audio, train_pinyin, mfcc, pinyin_dict, batch_size=batch_size, shuffle=True)
```

```python
# 定义转换函数。将音频文件提取特征，并进行转换
def convert(audio, feature_extr):
    feat, input_length = feature_extr.feature(audio, alignment=False)
    
    feat = feat[np.newaxis, :, :]
    feat = tf.convert_to_tensor(feat, tf.float32)
    input_length = tf.reshape(tf.convert_to_tensor([input_length], tf.int32), shape=[1,1])
    
    return feat, input_length
```

```python
audio = test_audio[0]
target_chinese = test_chinese[0]

feat, input_length = convert(audio, acoustic_audio_feature)

output, _ = acoustic_model.predict(feat, input_length, greedy=True)
pinyin_output = acoustic_pinyin_dict.decoding(output)

inputs = language_pinyin_dict.encoding(pinyin_output)
output, _ = language_model.predict(inputs, 
                                   y_start=language_chinese_dict.START, 
                                   y_end=language_chinese_dict.END)
chinese_output = language_chinese_dict.decoding(np.array(output))

print("Input: ", audio)
print("Target: ", target_chinese)
# print("Predict Pinyin: ", pinyin_output)
print("Predict Chinese: ", chinese_output)
```
